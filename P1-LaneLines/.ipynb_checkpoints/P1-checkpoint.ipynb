## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
# draw a line in image
def draw_linear_line(slope, intercept, x_coor, img, color=[255, 0, 0], thickness=10):
        
        
        height, width = img.shape[0:2]
        point_1 = (int(x_coor), int(slope * x_coor + intercept))
        
        if slope > 0:
            point_2 = (width, int(slope * width + intercept))
        else:
            point_2 = (0, int(intercept))

        # draw the line using cv2.line()
        cv2.line(img, point_1, point_2, color, thickness)
````

````python
# get the x coordinate of intersection of two lines
def get_intersection(pos_slope, pos_intercept, neg_slope, neg_intercept):
    x = (neg_intercept - pos_intercept) / (pos_slope - neg_slope)
    y = pos_slope * x + pos_intercept
    return x, y
````

````python
# given slope and intercept of all lines, then fit a line
def find_line_fit(slope_intercept, lines):
    slopes = [each[0] for each in slope_intercept]
    intercepts = [each[1] for each in slope_intercept]
    
    mean_slope = np.mean(slopes)
    std_slope = np.std(slopes)
    mean_intercept = np.mean(intercepts)
    new_lines = []
    
    
    # keep slope and intercept with more condition
    keep_slopes = []
    keep_intercepts = []
    
    for i in range(len(slope_intercept)):
        slope, intercept = slope_intercept[i]
        if slope - mean_slope < 1 * std_slope:
            keep_slopes.append(slope)
            keep_intercepts.append(intercept)
            new_lines.append(lines[i])
        else:
            pass
            #print("bingbing")
            #print(slope - mean_slope - 1 * std_slope)
    
    # update the mean slope and mean intercept with kepe lines
    if keep_slopes:
        mean_slope = np.mean(keep_slopes)
        mean_intercept = np.mean(keep_intercepts)
        lines = new_lines
    return mean_slope, mean_intercept, lines
````

````python
# do some processing to the original hough lines
def process_lines(lines, img_shape):
    pos_slope_intercept = []
    neg_slope_intercept = []
    pos_lines = []
    neg_lines = []
    width = img_shape[1]
    
    for line in lines:
        x_1, y_1, x_2, y_2 = line[0][0], line[0][1], line[0][2], line[0][3]
        if (x_1 - x_2 == 0):
            continue
        slope = (y_1 - y_2) / (x_1 - x_2)
        if np.abs(slope) < 0.09: # tan(5) = 0.87
            continue
        
        intercept = y_1 - slope * x_1
        dis = np.sqrt((x_1 - x_2)**2 + (y_1 - y_1)**2)
        
        if dis > 50:
            if slope > 0:
                pos_slope_intercept.append([slope, intercept])
                pos_lines.append(line)
            else:
                neg_slope_intercept.append([slope, intercept])
                neg_lines.append(line)
    
    if not pos_lines:
        for line in lines:
            x_1, y_1, x_2, y_2 = line[0][0], line[0][1], line[0][2], line[0][3]
            if (x_1 - x_2 == 0):
                continue
            slope = (y_1 - y_2) / (x_1 - x_2)
            if np.abs(slope) < 0.09: # tan(5) = 0.87
                continue
        
            intercept = y_1 - slope * x_1
        
            if slope > 0:
                pos_slope_intercept.append([slope, intercept])
                pos_lines.append(line)
     
    if not neg_lines:
        for line in lines:
            x_1, y_1, x_2, y_2 = line[0][0], line[0][1], line[0][2], line[0][3]
            if (x_1 - x_2 == 0):
                continue
            slope = (y_1 - y_2) / (x_1 - x_2)
            if np.abs(slope) < 0.09: # tan(5) = 0.87
                continue
        
            intercept = y_1 - slope * x_1
        
            if slope < 0:
                neg_slope_intercept.append([slope, intercept])
                neg_lines.append(line)
                
    if pos_lines:
        pos_slope, pos_intercept, new_pos_lines = find_line_fit(pos_slope_intercept, pos_lines)
    if neg_lines:
        neg_slope, neg_intercept, new_neg_lines = find_line_fit(neg_slope_intercept, neg_lines)\
    
    if pos_slope_intercept and neg_slope_intercept:
        # get the x coordinate of the intersection of two lines
        intersection_x_coor, intersection_y_coor = get_intersection(pos_slope, pos_intercept, neg_slope, neg_intercept)
        pos_x, pos_y = int(width), int(pos_slope * width + pos_intercept)
        neg_x, neg_y = 0, int(neg_intercept)
        
        proc_lines = new_pos_lines + new_neg_lines
        average_lines = [np.array([[int(intersection_x_coor), int(intersection_y_coor), pos_x, pos_y]]), np.array([[int(intersection_x_coor), int(intersection_y_coor), neg_x, neg_y]])]
    elif pos_slope_intercept:
        proc_lines = new_pos_lines
        x_1, y_1 = 0, int(pos_intercept)
        x_2, y_2 = int(width), int(pos_slope * width + pos_intercept)
        average_lines = [np.array([[x_1, y_1, x_2, y_2]])]
    else:
        proc_lines = new_neg_lines
        x_1, y_1 = 0, int(neg_intercept)
        x_2, y_2 = int(width), int(neg_slope * width + neg_intercept)
        average_lines = [np.array([[x_1, y_1, x_2, y_2]])]
    
    return proc_lines, average_lines
    
````

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=5):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)


def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    """
    ori_line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(ori_line_img, lines)
    
    average_line_img = average_lines(lines, img.shape)
    return ori_line_img, average_line_img
    """
    return lines
    
# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)
````

````python
# draw lane given image
def draw_lane_lines(img):
    plt.figure()
    plt.subplot(2, 2, 1)
    #plt.figure()
    plt.imshow(img)
    
    # grayscale the image and canny edge detection
    gray_img = grayscale(img)
    
    gray_img = gaussian_blur(gray_img, 5)
    
    low_thresh = 50  # threshold to supress one pixel as part of an edge
    hight_thresh = 150 # threshold to determine one pixel as part of an strong edge
    edge_img = canny(gray_img, low_thresh, hight_thresh)
    
    
    # apply the mask to an ROI, which can make us focus on ROI
    width = img.shape[1]
    height = img.shape[0]
    vertices = np.array([[[0, height],[440, 320], [520, 320], [width, height]]], dtype=np.int32)
    edge_img_with_mask = region_of_interest(edge_img, vertices)
    plt.subplot(2, 2, 2)
    #plt.figure()
    plt.imshow(edge_img_with_mask, cmap="gray")
    
    # use Hough Transform to detect line
    rho = 2 # distance resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 45     # minimum number of votes (intersections in Hough grid cell)
    min_line_len = 40 #minimum number of pixels making up a line
    max_line_gap = 100   # maximum gap in pixels between connectable line segments
    #ori_line_img, average_line_img = hough_lines(edge_img_with_mask, rho, theta, threshold, min_line_len, max_line_gap)
    
    lines = hough_lines(edge_img_with_mask, rho, theta, threshold, min_line_len, max_line_gap)
    proc_lines, average_lines = process_lines(lines, img.shape)
    
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, proc_lines)
    proc_color_img = weighted_img(line_img, img)
    plt.subplot(2, 2, 3)
    #plt.figure()
    plt.imshow(proc_color_img)
    
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, average_lines)
    line_img = region_of_interest(line_img, vertices)
    
    average_color_img = weighted_img(line_img, img)
    
    
    plt.subplot(2, 2, 4)
    #plt.figure()
    plt.imshow(average_color_img)
    
draw_lane_lines(imgs[0])    
````

### Pipeline description
#### pipeline descriptions
1. Read in and grayscale image
2. Define a kernel size and apply Gaussian smoothing 
3. Define parameters for Canny edge detection and apply to get edge image
4. Mask edge image using cv2.fillyPoly(ignore everything out of region of interest), make the edge image more clean for Hough line detection
5. Define Hough Transform parameters, run it on masked edge image, then get a collection of line segments
6. Draw line segments
7. Filter line segments not belonging to the lane line and average remaining line segments to get straight lane lines(left one and right one)
8. Combine remaining lane lines and original image to see how accurate the predicted lane lines are 

#### Focus on step 7
The most hard part is to get correct lane line segments and average(or extrapolate) to get complete left and right lane lines, below is what I refer to
1. Divide all line segments into two group, one with positive slope(backslash), the other with negative slope(forward slash)
2. Filter each group by the slope mean and slope standard deviation, then get updated line segments for each group 3. Use the mean slope and mean intercept of each updated group as the final lane line slope and intercept
4. Draw the lane lines from intersection point to the bottom of the image

### Shortcomings 
1. In solidYellowLeft video, in some frames, the resulted right lane **diverges to left**, there are some **horizontal line segments** from Hough Transform, which affect the following filtering process. 
2. The ROI mask is designed by hand, but in real time, the mask should be related to the postion of camera. In the challeng video, the hard-coded mask includes the left road edge, then there are unexpected line segments
3. In the challenge video,when the lane line has high curvature, the results are really bad
4. The algorithm is designed for lane line and specific type of day (daytime, moderately bright with no snow, rain or hail), so it may not generalize well

### Possible improvements
1. change the line segments filtering method for each group(positive and negative)
2. improvements on curved lane line
3. imporvements for changed car and camera position
